# **Checkpoint 5 : Sector Analysis 2**

```python
#Creating D1 for USA (Country 1)
D1_mf = master_frame[(master_frame['country_code'] == 'USA') & 
             (master_frame['raised_amount_usd'] >= 5000000) & 
             (master_frame['raised_amount_usd'] <= 15000000)]
D1_usa = D1_mf[['raised_amount_usd','main_sector']].groupby('main_sector').agg(['sum', 'count']).rename(
    columns={'sum':'total_amount','count' : 'total_number'})
D1 = pd.merge(D1_mf, D1_usa, how='left', on='main_sector')
D1.head()
```

```python
#Creating D2 for GBR (Country 2)
D2_mf = master_frame[(master_frame['country_code'] == 'GBR') & 
             (master_frame['raised_amount_usd'] >= 5000000) & 
             (master_frame['raised_amount_usd'] <= 15000000)]
D2_gbr = D2_mf[['raised_amount_usd','main_sector']].groupby('main_sector').agg(['sum', 'count']).rename(
    columns={'sum':'total_amount','count' : 'total_number'})
D2 = pd.merge(D2_mf, D2_gbr, how='left', on='main_sector')
D2.head()
```

```python
#Creating D3 for IND (Country 3)
D3_mf = master_frame[(master_frame['country_code'] == 'IND') & 
             (master_frame['raised_amount_usd'] >= 5000000) & 
             (master_frame['raised_amount_usd'] <= 15000000)]
D3_ind = D3_mf[['raised_amount_usd','main_sector']].groupby('main_sector').agg(['sum', 'count']).rename(
    columns={'sum':'total_amount','count' : 'total_number'})
D3 = pd.merge(D3_mf, D3_ind, how='left', on='main_sector')
D3.head()
```

```python
#Total amount of investment (USD)
print(round(D1.raised_amount_usd.sum(), 2))
print(round(D2.raised_amount_usd.sum(), 2))
print(round(D3.raised_amount_usd.sum(), 2))
```

# **Checkpoint 6 : Plots** 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Plotting plot 1.1 : fraction of total investments (globally) in venture, seed, and private equity
master_frame_plotting = master_frame_plotting[~(master_frame_plotting['funding_round_type'] =='angel')]
plt.figure(figsize=(12, 8))
sns.countplot(x="funding_round_type", data=master_frame_plotting)
plt.title("Fraction of Total Investments")
plt.show()
```

```python
#Plotting plot 1.2 : Average amount of investment in venture, seed, and private equity
plt.figure(figsize=(12, 8))
sns.barplot(y='raised_amount_usd', x="funding_round_type", data=master_frame_plotting, estimator=np.mean)
plt.title("Average amount of Investment")
plt.ylabel('Amount Invested (in 10 million USD)')
plt.axhline(y=5000000, linewidth=2, color = 'r')
plt.axhline(y=15000000, linewidth=2, color = 'r')
plt.show()
```

```python
# Plotting plot 2 : top 9 countries against the total amount of investments of funding type 
top9_plotting['country_code'] = top9_plotting.index
plt.figure(figsize=(12, 8))
sns.barplot(x="country_code", y="raised_amount_usd", data=top9_plotting)
plt.title("Sum Total Investment")
plt.show()
```

```python
#Understanding top 3 sectors 1st for USA
plt.figure(figsize=(20, 5))
sns.barplot(x="main_sector", y="raised_amount_usd", data=D1, estimator=sum)
plt.title("Sum")
plt.show()
D1_plot3 = D1[D1['main_sector'].isin(["Others","Social, Finance, Analytics, Advertising", "Cleantech / Semiconductors"])]
```

```python
#Understanding top 3 sectors 1st for GBR
plt.figure(figsize=(20, 5))
sns.barplot(x="main_sector", y="raised_amount_usd", data=D2, estimator=sum)
plt.title("Sum")
plt.show()
D2_plot3 = D2[D2['main_sector'].isin(["Others","Social, Finance, Analytics, Advertising","Cleantech / Semiconductors"])]
```

```python
#Understanding top 3 sectors 1st for IND
plt.figure(figsize=(20,5))
sns.barplot(x="main_sector", y="raised_amount_usd", data=D3, estimator=sum)
plt.title("Sum")
plt.show()
D3_plot3 = D3[D3['main_sector'].isin(["Others","Social, Finance, Analytics, Advertising","News, Search and Messaging"])]
```

```python
#Plotting Fig 3 : No of investmentments in top3 countires in top3 sectors
frames = [D1_plot3, D2_plot3, D3_plot3]
result = pd.concat(frames)
plt.figure(figsize=(12,8))
plt.ylabel('No of Investments(in 1000)')
sns.barplot(x='country_code', y='raised_amount_usd', hue="main_sector", data=result, estimator=np.sum)
plt.show()
```
